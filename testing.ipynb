```python
#Making histogram that shows accuracy of current genre predicting algorithm

low_index=0  #index bounds for Dark Trap
high_index=4577
Song_names = np.loadtxt(
    "genre_predictor/data/genres_v2.csv",
    dtype=str,
    skiprows=1,
    delimiter=",",
    usecols=(19),
    encoding="utf8",
)



def genre_tester(high_index:int,low_index:int):
    dark_trap_counter=0
    trap_metal_counter=0
    und_counter=0
    emo_counter=0
    rap_counter=0
    rnb_counter=0
    pop_counter=0
    hiphop_counter=0
    techhouse_counter=0
    techno_counter=0
    trance_counter=0
    psytrance_counter=0
    trap_counter=0
    dnb_counter=0
    Hardstyle_counter=0

    for i in range(high_index-low_index):
        song_name=Song_names[i]
        genre_predicted = Genre_Predictor(song_name)
        
        if genre_predicted=='Dark Trap':
            dark_trap_counter+=1

        if genre_predicted=='Underground Rap':
            und_counter+=1   

        if genre_predicted=='Trap Metal':
            trap_metal_counter+=1

        if genre_predicted=='Emo':
            emo_counter+=1

        if genre_predicted=='Rap':
            rap_counter+=1

        if genre_predicted=='RnB':
            rnb_counter+=1

        if genre_predicted=='Pop':
            pop_counter+=1

        if genre_predicted=='Hiphop':
            hiphop_counter+=1

        if genre_predicted=='techhouse':
            techhouse_counter+=1    

        if genre_predicted=='techno':
            techno_counter+=1

        if genre_predicted=='trance':
            trance_counter+=1

        if genre_predicted=='psytrance':
            psytrance_counter+=1

        if genre_predicted=='trap':
            trap_counter+=1

        if genre_predicted=='dnb':
            dnb_counter+=1

        if genre_predicted=='hardstyle':
            Hardstyle_counter+=1   

    genre_counter=np.array([dark_trap_counter,und_counter, trap_metal_counter, emo_counter,rap_counter,rnb_counter,pop_counter,hiphop_counter,techhouse_counter,techno_counter,trance_counter,psytrance_counter,trap_counter,dnb_counter,Hardstyle_counter])
    return genre_counter


genre_names = [
    "DarkTrap",
    "Und_rap",
    "TrapMetal",
    "Emo",
    "Rap",
    "RnB",
    "Pop",
    "Hiphop",
    "Techhouse",
    "Techno",
    "Trance",
    "Psytrance",
    "Trap",
    "Dnb",
    "Hardstyle",
]


```
